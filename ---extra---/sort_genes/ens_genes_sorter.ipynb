```python
import pandas as pd
import numpy as np
```

```python
hgnc = pd.read_csv("hgnc_complete_set_2022-07-01.txt", sep="\t").dropna(subset=["Chromosome", "Ensembl gene ID"])
hgnc = hgnc[["Approved symbol", 'Approved name', 'Status', 'Chromosome', 'Ensembl gene ID']]
hgnc
```

```python
hgnc["Mitochondria"] = hgnc.apply(
    lambda row: "yes" if row["Chromosome"] == "mitochondria" else "no"
    , axis=1)
hgnc = hgnc[hgnc["Chromosome"] != "mitochondria"]
hgnc = hgnc[hgnc["Status"] != "Symbol Withdrawn"]
hgnc = hgnc[hgnc["Status"] != "Entry Withdrawn"]
hgnc
```

```python
if "Chromosome" in hgnc:
    
    hgnc["arm_pos"] = ""
    
    hgnc["arm_pos"] = hgnc.apply(
        lambda row: row["Chromosome"].find("p") if "p" in row["Chromosome"] else row["Chromosome"].find("q")
        , axis=1)
    
    
    hgnc["chr"] = hgnc.apply(
        lambda row: row['Chromosome'][:row["arm_pos"]]
        , axis=1)
    
    hgnc["chr"] = hgnc.apply(
        lambda row: str(row['chr']).zfill(2) if row['chr'] not in ["X", "Y"] else row['chr']
        , axis=1)
    
    hgnc["arm"] = hgnc.apply(
        lambda row: row["Chromosome"][row["arm_pos"]]
        , axis=1)
    
    hgnc["pos"] = ""
    
    hgnc["pos"] = hgnc.apply(
        lambda row: row["Chromosome"][row["arm_pos"] + 1:]
        , axis=1)
    

    #del hgnc["Chromosome"]
    del hgnc["arm_pos"]

hgnc
```

```python
hgnc.sort_values(["chr", "arm", "pos"], ascending=True, inplace=True)
hgnc.reset_index(inplace=True)
del hgnc["level_0"]
hgnc
```

```python
del hgnc["index"]
hgnc.to_csv("hgnc_complete_set_2022-07-01.csv", index=False)
```

```python
df = pd.read_csv("data.csv", index_col = 0)
df
```

```python
df_genes = list(df.columns)
len(df_genes)
```

```python
df_sorted_genes = []
#after we sorted the ref df, lets sort our genes
for index, row in hgnc.iterrows():
    enst_id = row["Ensembl gene ID"]
    if enst_id in df_genes:
        df_sorted_genes.append(enst_id)
        df_genes.remove(enst_id)
        
len(df_sorted_genes), len(df_genes)
```

```python
df_sorted_genes
```

```python
df_new_cols = df_sorted_genes + df_genes
len(df_new_cols)
```

```python
df_sorted = df[df_new_cols]
df_sorted
```

```python
df_sorted.to_csv("data_sorted.csv")
```
